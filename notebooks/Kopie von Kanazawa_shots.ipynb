```python
RATE = 10
VID_NO = 11
```

```python
CLASSES = ["Gedan-Barai",
"Oi-Tzuki","Transition"]
```

```python
# Read the video from specified path
# cam = cv2.VideoCapture("/content/Heian Shodan - Shotokan (SKIF) - Hirokazu Kanazawa [lZyI9mtk924].mp4")#"/content/Jion - Shotokan (SKIF) - Hirokazu Kanazawa [cQdAZ94OkNA].mp4")
cam = cv2.VideoCapture(os.path.join("/content/",filenames[VID_NO]))

print("framerate:",cam.get(cv2.CAP_PROP_FPS))

try:
    dir_path = ""
    # creating a folder named data
    for cls in CLASSES:
        dir_path = os.path.join('/content/drive/MyDrive/K/Kanazawa/data/train',cls)
        if not os.path.exists(dir_path):
            os.makedirs(dir_path)

    # if not created then raise error
except OSError:
    print(f'Error: Creating directory {dir_path} failed')

# frame
currentframe = 0

clicked = False

def on_button0_clicked(b):
    global clicked
    clicked = True
    # if video is still left continue creating images
    name = '/content/drive/MyDrive/K/Kanazawa/data/train/Gedan-Barai/frame_'+str(VID_NO)+'_' + str(currentframe//100).zfill(4) + '.jpg'
    cv2.imwrite(name, frame)

def on_button1_clicked(b):
    global clicked
    clicked = True
    # if video is still left continue creating images
    name = '/content/drive/MyDrive/K/Kanazawa/data/train/Oi-Tzuki/frame_'+str(VID_NO)+'_' + str(currentframe//100).zfill(4) + '.jpg'
    cv2.imwrite(name, frame)

def make_callback(cls):
    def on_button_clicked(b):
        global clicked
        clicked = True
        # if video is still left continue creating images
        name = '/content/drive/MyDrive/K/Kanazawa/data/train/' + cls + '/frame_' + str(VID_NO) + '_' + str(currentframe//100).zfill(4) + '.jpg'
        cv2.imwrite(name, frame)
    return on_button_clicked

def on_button_skip_clicked(b):
    global clicked
    clicked = True

output = widgets.Output()
buttons = []
for cls in CLASSES:
    button = widgets.Button(description=cls)
    button.on_click(make_callback(cls))
    buttons.append(button)
#button0 = widgets.Button(description="Gedan-Barai")
#button1 = widgets.Button(description="Oi-Tzuki")
button_skip = widgets.Button(description="skip")
#button0.on_click(on_button0_clicked)
#button1.on_click(on_button1_clicked)
button_skip.on_click(on_button_skip_clicked)
display(output, *buttons, button_skip)

with ui_events() as poll:
    while (True):
        # time.sleep(60) # take schreenshot every 5 seconds
        # reading from frame
        ret, frame = cam.read()

        if ret:
            if currentframe%RATE==0:
                # if video is still left continue creating images
                name = './data/frame' + str(currentframe//RATE).zfill(4) + '.jpg'
                # print('Creating...' + name)
                with output:
                    clear_output(wait=True)
                    cv2_imshow(frame)
                clicked = False
                while not clicked:
                    poll(10) # poll queued UI events including button
                    time.sleep(0.1) # wait for 1 second before checking again
                # writing the extracted images
                cv2.imwrite(name, frame)

            # increasing counter so that it will
            # show how many frames are created
            currentframe += 1
        else:
            break

# Release all space and windows once done
cam.release()
cv2.destroyAllWindows()
```
